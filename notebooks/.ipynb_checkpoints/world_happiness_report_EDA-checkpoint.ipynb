## Top 10 happiest countries per year

```python
years = df["Year"].unique()
years.sort()
years
```

```python
for year in years:
    data = df[df["Year"] == year].groupby("Country Name")["Life Ladder"]\
    .mean().round(2).nlargest(10).reset_index()
    fig, ax = plt.subplots(figsize=(10, 5))
    sns.set(style = "whitegrid")
    palette = sns.color_palette("Blues", n_colors=10)
    palette.reverse()
    sns.barplot(ax=ax, data=data, x=data["Life Ladder"], y=data["Country Name"], palette=palette)
    ax.set_title(f"Top 10 Happiest Countries of {year}")
    ax.bar_label(ax.containers[0])
```
